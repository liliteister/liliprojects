### Geographic Data
Links and basic code to download geojson data from Portland Maps into geopandas dataframes.


----

```python
import geopandas as gpd
```

Download data from Portland Maps containing parks, farmers markets, and community gardens in Portland, OR. They are available as geojson files.

- [Parks](https://gis-pdx.opendata.arcgis.com/datasets/parks?geometry=-123.123%2C45.457%2C-122.080%2C45.625)
- [Farmer's Markets](https://gis-pdx.opendata.arcgis.com/datasets/farmers-markets)
- [Community Gardens](https://gis-pdx.opendata.arcgis.com/datasets/community-gardens?geometry=-123.157%2C45.450%2C-122.114%2C45.618)

```python
parks_link = "https://opendata.arcgis.com/datasets/9eef54196eaa4d12b54e9bc40e70ff09_35.geojson"
farm_link = "https://opendata.arcgis.com/datasets/91918d35588345a4b0f59f246e6e3866_187.geojson"
gard_link = "https://opendata.arcgis.com/datasets/94e5a7a096fb41c2bc38f419042ee9b5_216.geojson"

PDX_parks = gpd.read_file(parks_link)
PDX_farmers_markets = gpd.read_file(farm_link)
PDX_comm_gardens = gpd.read_file(gard_link)
```

```python
PDX_parks.plot()
PDX_farmers_markets.plot()
PDX_comm_gardens.plot()  #they are really small
```
